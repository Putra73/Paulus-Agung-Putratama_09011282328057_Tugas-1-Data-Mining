```python
Kolom = df[['id', 'age', 'trestbps', 'chol', 'thalch', 'oldpeak', 'ca', 'num']]
Korelasi = Kolom.corr()
print(Korelasi)
```

```python
Kolom = df[['id', 'age', 'trestbps', 'chol', 'thalch', 'oldpeak', 'ca', 'num']]
for Kol in Kolom:
    plt.figure()
    df[Kol].hist(bins= 20)
    plt.title(f'Histogram of {Kol}')
    plt.xlabel(Kol)
    plt.ylabel('Frequency')
    plt.show()
```

```python
Kolom = df[['id', 'age', 'trestbps', 'chol', 'thalch', 'oldpeak', 'ca', 'num']]
for Kol in Kolom:
    plt.figure()
    sns.boxplot(x=df[Kol])
    plt.title(f'Boxplot of {Kol}')
    plt.show()
```

```python
Kolom = df[['id', 'age', 'trestbps', 'chol', 'thalch', 'oldpeak', 'ca', 'num']]
Korelasi = Kolom.corr()

columns = Korelasi.columns
upper_triangle_matrix = Korelasi.where(np.triu(np.ones(Korelasi.shape), k=1).astype(bool))
max_correlation = upper_triangle_matrix.max().max()
max_correlation_pair = upper_triangle_matrix.stack().idxmax()

print(f'Pasangan kolom dengan korelasi tertinggi: {max_correlation_pair} dengan nilai korelasi: {max_correlation}')
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df[max_correlation_pair[0]], df[max_correlation_pair[1]], alpha=0.7)
plt.title(f'Scatter Plot of {max_correlation_pair[0]} vs {max_correlation_pair[1]}')
plt.xlabel(max_correlation_pair[0])
plt.ylabel(max_correlation_pair[1])
plt.show()
```
